### 5. **Data Preprocessing**


```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split,KFold
from sklearn.preprocessing import LabelEncoder,PolynomialFeatures
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
train_red['train']  = 1
test_red['train']  = 0
```

```python
data = pd.concat([train_red.drop('Amount (Million Naira)',axis  =  1), test_red]).reset_index(drop = True)
data.head()
```

```python
##before
data[['Year', 'Maker', 'Model', 'Location', 'Type']].sample(5)
```

```python
##after
''.join(data[['Year', 'Maker', 'Model', 'Location', 'Type']].astype(str).iloc[0].values)
```

```python
mashup = [''.join(data[['Year', 'Maker', 'Model', 'Location', 'Type']].astype(str).iloc[i].values) for i in range(len(data))]
### 
mashup_cleaned = [mashup[i].strip().replace(',','') for i in range(len(mashup))]

Vectorizer= CountVectorizer(
                analyzer = "char",  
                tokenizer = None,  
                preprocessor = None, 
                stop_words = None,  
                max_features = 5000,
)
Vectorizer.fit(mashup_cleaned)
alpha_to_array = Vectorizer.transform(mashup_cleaned).toarray()
vocab = Vectorizer.vocabulary_
word_vec_df = pd.DataFrame(alpha_to_array,columns=vocab)
```

```python
word_vec_df.head()
```

```python
cat_col = ['Maker', 'Model', 'Type']

for col in cat_col:
    lab = LabelEncoder()#CountEncoder(normalize=True)
    lab.fit(data[col])
    data[col] = lab.transform(data[col])
```

```python
data['Year'] = data['Year'].str.replace(',','')
```

### 6. **Modelling**


```python
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor, XGBRFRegressor
from sklearn.tree import DecisionTreeRegressor
from catboost import CatBoostRegressor
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor,BaggingRegressor,VotingRegressor,StackingRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
#Try other models to figure out which one gives the best score using GridSearchCV
from sklearn.model_selection import ShuffleSplit, cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
#from catboost import CatBoostRegressor

#We now write a function

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        "LGBM":{
            "model":LGBMRegressor(),
            "params":{
                "num_leaves":[21, 31, 41],
                "max_depth": [1, 2, -1,],
                "learning_rate": [0.1, 0.01],
                "n_estimators": [10, 100]
               
            }
        },
        
        "XGBoost": {
            "model": XGBRegressor(),
            "params": {
                "max_depth":[1, 2, -1],
                "learning_rate":[0.1, 0.01, 0.001]
                
            }
        },
                
        "RandomForest": {
            "model": RandomForestRegressor(),
            "params": {
                'bootstrap': [False],
                'n_estimators': [3, 10],
                'max_features': [2, 3, 4]
                
            }
        },
        
        "Decision_Tree": {
            "model": DecisionTreeRegressor(),
            "params": {
                "criterion": ["mse", "friedman_mse"],
                "splitter": ["best", "random"]
            }
        }

    }
    #To save the model into a dataframe
    
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config["model"], config["params"], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            "model":algo_name,
            "best_score":gs.best_score_,
            "best_params":gs.best_params_
        })
        
    return pd.DataFrame(scores, columns=["model","best_score", "best_params"])

# Calling the function

find_best_model_using_gridsearchcv(features, labels)
    
```
